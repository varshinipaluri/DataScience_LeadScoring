```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('selected_features.csv')

df.drop(['Prospect ID'], axis=1 , inplace=True)
df.drop(['Asymmetrique Activity Index', 'Asymmetrique Profile Index'], axis=1, inplace=True)
```

```python
df=df.iloc[:,0:14]
# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Define number of rows and columns for subplots
num_cols = df.shape[1]
cols_per_row = 3  # 3 plots per row
rows = -(-num_cols // cols_per_row)  # Ceiling division

# Create subplots
fig, axes = plt.subplots(rows, cols_per_row, figsize=(16, 5 * rows))
axes = axes.flatten()  # Flatten to loop easily

# Plot histograms with unique colors
colors = sns.color_palette("husl", num_cols)  # Unique color for each plot

for i, col in enumerate(df.columns):
    sns.histplot(df[col], bins=20, color=colors[i], ax=axes[i], edgecolor='black')
    axes[i].set_title(col, fontsize=14, fontweight='bold')
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Improve layout and add title
plt.suptitle('Enhanced Data Distribution for All Columns', fontsize=18, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
